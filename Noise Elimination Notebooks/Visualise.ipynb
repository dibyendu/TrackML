`[<dict>, <dict>, ... 100 ..., <dict>]` $\leftrightarrow$ `*.json`

 
                                          <dict>
                                            |
                       ----------------------------------------------
                       |                                            |
                       hL                                          gthL
                 [6 x <set>]                                    [6 x <set>]

`<set> := list of N unique points of the form (x,y)`

E.g: for $N=3$,

$[(27, 950), (35, 352), (35, 353)]$

```python
PATH_Prefix = '../'
DATA_DIR = 'InclusiveEvents_Iteration1_Parsed/'
data_file_path = PATH_Prefix + DATA_DIR + 'dataset_0002.json'
data_point = 0
```

```python
import sys
sys.path.append(PATH_Prefix)

import ujson, numpy as np
from plotLayer import PlotImage, PlotSingleImage
```

```python
with open(data_file_path, 'r') as f:
    input_raw_data = ujson.loads(f.read())
```

```python
PlotSingleImage(input_raw_data[data_point])
# PlotSingleImage(input_raw_data[data_point], width=512)
```

```python
input_hl_data, input_gthl_data = [], []
for hL in input_raw_data[data_point]['hL']:
    zeros = np.zeros((1024, 1024))
    zeros[tuple(np.array(list(hL)).T.tolist())] = 1
    input_hl_data.append(zeros)

for gthL in input_raw_data[data_point]['gthL']:
    zeros = np.zeros((1024, 1024))
    zeros[tuple(np.array(list(gthL)).T.tolist())] = 1
    input_gthl_data.append(zeros)

PlotImage(np.array(input_hl_data), np.array(input_gthl_data))
```
